```python
import requests
import html_to_json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import selenium
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
import re
import pandas as pd
```

```python
homepage = 'https://vegetablemarketprice.com/'
data = []
html_list = []
driver = webdriver.Firefox()
driver.get(homepage)
search_input = driver.find_element(by=By.XPATH, value='//*[@id="header-search-input"]')

header_search_input = 'Delhi Vegetable Market Price History'
search_input.send_keys(header_search_input)
search_input.send_keys(Keys.ENTER)
time.sleep(5)
driver.find_element(by=By.XPATH,value='/html/body/ul/li[1]/div').click()


switch = driver.find_element(By.XPATH, '//*[@id="navItemTable"]')
switch.click()

change_date = driver.find_element(by = By.XPATH,value = '//*[@id="reportrange"]')

change_date.click()
custom_range = driver.find_element(By.XPATH, "//*[contains(text(), 'Custom Range')]")
custom_range.click()
```

```python
# Initialzing the dates
while True: 
    check_month = driver.find_element(By.XPATH, "/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[2]")
    month = check_month.get_attribute("innerHTML")
    if month == "Feb 2022":
        break
        
    else:
        back_button = driver.find_element(By.XPATH,'/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[1]')

        back_button.click()
        
first_date = driver.find_element(By.XPATH, '/html/body/div[4]/div[2]/div[1]/table/tbody/tr[1]/td[3]')
first_date.click()
```

```python
# selecting date from feb to july
i=0
while i <= 2:
    front_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/thead/tr[1]/th[3]")
    front_button.click()
    i = i+1

last_date = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/tbody/tr[5]/td[7]")
last_date.click()
```

```python
apply_button = driver.find_element(By.XPATH, '/html/body/div[4]/div[4]/button[2]')
driver.execute_script("arguments[0].scrollIntoView();", apply_button)
apply_button.click()
```

```python
final_list = ['Onion Big', 'Onion Small', 'Tomato', 'Green Chilli', 'Beetroot', 'Potato',
    'Raw Banana (Plantain)', 'Amaranth Leaves', 'Amla', 'Ash gourd', 'Baby Corn',
    'Banana Flower', 'Capsicum', 'Bitter Gourd', 'Bottle Gourd', 'Butter Beans',
    'Broad Beans', 'Cabbage', 'Carrot', 'Cauliflower', 'Cluster beans', 'Coconut',
    'Colocasia Leaves', 'Colocasia', 'Coriander Leaves', 'Corn', 'Cucumber',
    'Curry Leaves', 'Dill Leaves', 'Drumsticks', 'Brinjal', 'Brinjal (Big)',
    'Elephant Yam', 'Fenugreek Leaves', 'French Beans', 'Garlic', 'Ginger',
    'Onion Green', 'Green Peas', 'Ivy Gourd', 'Lemon (Lime)', 'Mango Raw',
    'Mint Leaves', 'Mushroom', 'Mustard Leaves', 'Ladies Finger', 'Pumpkin',
    'Radish', 'Ridge Gourd', 'Shallot (Pearl Onion)', 'Snake Gourd', 'Sorrel Leaves',
    'Spinach', 'Sweet Potato']

```

```python
for i in range (0,len(final_list)):
    while True:
        try:
            element = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div[1]/div[4]/div/div[1]/div/div[1]/button')
            element.click()
        except:
            break
    search_input = driver.find_element(By.NAME, 'search_terms')
    search_input.send_keys(final_list[i])
    time.sleep(3)
    search_input.send_keys(Keys.ENTER)
    
    temp = driver.find_element(By.XPATH, '//*[@id="historyDataTable"]')
    time.sleep(3)
    temp_html = temp.get_attribute("innerHTML")
    html_list.append(temp_html)
    time.sleep(3)
    print(final_list[i])
```

```python
date = re.findall('<th>(.*?)</th>',html_list[0])
dates = []
for i in date:
    temp = re.findall('"white-space: nowrap;">(.*?)</span>',i)
    if len(temp) != 0:
        dates.append(temp)
```

```python
Name = []
Prices = []

for i in html_list:
    temp_n = re.findall('<tbody><tr><td>(.*?)</td><td>',i)
    Name.extend(temp_n)
    
    temp_p = re.findall('<td>₹ (.*?)</td>',i)
    Prices.append(temp_p)
```

```python
#converting into dataframe 
df = pd.DataFrame(dates,columns = ['Dates'])
df1 = pd.DataFrame(Prices)
df1 = df1.transpose()
df1.dropna(inplace = True)
df1.columns = [Name]
df = pd.merge(df,df1,left_index=True, right_index=True)
```

## Extracting data from aug to dec

```python
#visiting to noida market price history from main page
homepage = 'https://vegetablemarketprice.com/'
data = []
html_list = []
driver = webdriver.Firefox()
driver.get(homepage)
search_input = driver.find_element(by=By.XPATH, value='//*[@id="header-search-input"]')

header_search_input = 'Delhi Vegetable Market Price History'
search_input.send_keys(header_search_input)
search_input.send_keys(Keys.ENTER)
time.sleep(5)
driver.find_element(by=By.XPATH,value='/html/body/ul/li[1]/div').click()

#selecting table in dashboard
switch = driver.find_element(By.XPATH, '//*[@id="navItemTable"]')
switch.click()

#selecting custom range from the date 
change_date = driver.find_element(by = By.XPATH,value = '//*[@id="reportrange"]')
change_date.click()
custom_range = driver.find_element(By.XPATH, "//*[contains(text(), 'Custom Range')]")
custom_range.click()


# Initialzing the dates
while True: 
    check_month = driver.find_element(By.XPATH, "/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[2]")
    month = check_month.get_attribute("innerHTML")
    if month == "Aug 2022":
        break
        
    else:
        back_button = driver.find_element(By.XPATH,'/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[1]')

        back_button.click()
        
first_date = driver.find_element(By.XPATH, '/html/body/div[4]/div[2]/div[1]/table/tbody/tr[1]/td[2]')
first_date.click()

#selecting date from Aug to Dec
i=0
while i <= 2:
    front_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/thead/tr[1]/th[3]")
    front_button.click()
    i = i+1

last_date = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/tbody/tr[5]/td[7]")
last_date.click()

apply_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[4]/button[2]")
apply_button.click()
```

```python
#Extracting  the names of vegetables 
for i in range (0,len(final_list)):
    while True:
        try:
            element = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div[1]/div[4]/div/div[1]/div/div[1]/button')
            element.click()
        except:
            break
    search_input = driver.find_element(By.NAME, 'search_terms')
    search_input.send_keys(final_list[i])
    time.sleep(3)
    search_input.send_keys(Keys.ENTER)
    
    temp = driver.find_element(By.XPATH, '//*[@id="historyDataTable"]')
    time.sleep(3)
    temp_html = temp.get_attribute("innerHTML")
    html_list.append(temp_html)
    time.sleep(3)
    print(final_list[i])

#extracting names of vegetables and prices
Name = []
Prices = []

for i in html_list:
    temp_n = re.findall('<tbody><tr><td>(.*?)</td><td>',i)
    Name.extend(temp_n)
    
    temp_p = re.findall('<td>₹ (.*?)</td>',i)
    Prices.append(temp_p)

#extracting date 
date = re.findall('<th>(.*?)</th>',html_list[0])
dates = []
for i in date:
    temp = re.findall('"white-space: nowrap;">(.*?)</span>',i)
    if len(temp) != 0:
        dates.append(temp)
```

```python
# converting into Dataframe
df = pd.DataFrame(dates,columns = ['Dates'])
df1 = pd.DataFrame(Prices).transpose()
df1.dropna(inplace = True)
df1.columns = [Name]
df = pd.merge(df,df1,left_index=True, right_index=True)
```

```python
# exporting the data into csv file
p='E:\\Masai DA Projects\\AUG_DEC_Prices_DELHI.csv'
df.to_csv(p, index=False)
```
