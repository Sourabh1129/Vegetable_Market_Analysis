extracting list of all vegetables 

```python
link = "https://vegetablemarketprice.com/market/noida/today"

driver = webdriver.Firefox()
driver.get(link)
veges = driver.find_element(By.XPATH, '//*[@id="todayVegetableTableContainer"]')
vege = veges.get_attribute("innerHTML")

vege_h_list = re.findall('<td>(.*?)</td>',vege)
final_list = []
size = len(vege_h_list) 
for i in range (size):
    if i%5 == 0:
        final_list.append(vege_h_list[i])
```

Extracting dates and prices of vegetables

```python
# opening a firefox browser and visiting to site 
homepage = 'https://vegetablemarketprice.com/'
data = []
html_list = []
driver = webdriver.Firfox()
driver.get(homepage)
search_input = driver.find_element(by=By.XPATH, value='//*[@id="header-search-input"]')

#searching for Noida Market Price History 
header_search_input = 'Noida Vegetable Market Price History'
search_input.send_keys(header_search_input)
search_input.send_keys(Keys.ENTER)
time.sleep(5)
driver.find_element(by=By.XPATH,value='/html/body/ul/li[1]/div').click()

# # switching from charts to table in dashboard
switch = driver.find_element(By.XPATH, '//*[@id="navItemTable"]')
switch.click()

#customizing the date and selecting custom date a
change_date = driver.find_element(by = By.XPATH,value = '//*[@id="reportrange"]')

change_date.click()
custom_range = driver.find_element(By.XPATH, "//*[contains(text(), 'Custom Range')]")
custom_range.click()

# initializing the start date
while True: 
    check_month = driver.find_element(By.XPATH, "/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[2]")
    month = check_month.get_attribute("innerHTML")
    if month == "Feb 2022":
        break
        
    else:
        back_button = driver.find_element(By.XPATH,'/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[1]')

        back_button.click()
        
first_date = driver.find_element(By.XPATH, '/html/body/div[4]/div[2]/div[1]/table/tbody/tr[1]/td[3]')
first_date.click()

# switching to the next page of calender 
i = 0
while i <= 3:
    front_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/thead/tr[1]/th[3]")
    front_button.click()
    i = i+1

last_date = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/tbody/tr[5]/td[7]")
last_date.click()

#applying the custom date ( feb - july)
apply_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[4]/button[2]")
apply_button.click()

# #Extracting  the names of vegetables available 
for i in range (0,len(final_list)):
    while True:
        try:
            element = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div[1]/div[4]/div/div[1]/div/div[1]/button') #removing the default names from search inputs 
            element.click()
        except:
            break
# searching the prices of each vegetable from custom dates
    search_input = driver.find_element(By.NAME, 'search_terms') 
    search_input.send_keys(final_list[i])
    
    search_input.send_keys(Keys.ENTER)
    
    temp = driver.find_element(By.XPATH, '//*[@id="historyDataTable"]')
    
    temp_html = temp.get_attribute("innerHTML")
    html_list.append(temp_html)
    time.sleep(10)

    print(final_list[i])

```

```python
date = re.findall('<th>(.*?)</th>',html_list[0])
dates = []
for i in date:
    temp = re.findall('"white-space: nowrap;">(.*?)</span>',i)
    if len(temp) != 0:
        dates.append(temp)
```

```python
df = pd.DataFrame(dates,columns = ['Dates'])
```

```python
Name = []
Prices = []

for i in html_list:
    temp_n = re.findall('<tbody><tr><td>(.*?)</td><td>',i)
    Name.append(temp_n)
    
    temp_p = re.findall('<td>₹ (.*?)</td>',i)
    Prices.append(temp_p)
```

```python
#converting into dataframe 
df1 = pd.DataFrame(Prices)
df1 = df1.transpose()
df1.dropna(inplace = True)
df1.columns = [Name]
df = pd.merge(df,df1,left_index=True, right_index=True)
```

```python
# visiting to noida market price history from main page
homepage = 'https://vegetablemarketprice.com/'
data = []
html_list = []
driver = webdriver.Chrome()
driver.get(homepage)
search_input = driver.find_element(by=By.XPATH, value='//*[@id="header-search-input"]')

header_search_input = 'Noida Vegetable Market Price History'
search_input.send_keys(header_search_input)
search_input.send_keys(Keys.ENTER)
time.sleep(5)
driver.find_element(by=By.XPATH,value='/html/body/ul/li[1]/div').click()

#selecting table in dashboard
switch = driver.find_element(By.XPATH, '//*[@id="navItemTable"]')
switch.click()

#selecting custom range from the date 
change_date = driver.find_element(by = By.XPATH,value = '//*[@id="reportrange"]')
change_date.click()
custom_range = driver.find_element(By.XPATH, "//*[contains(text(), 'Custom Range')]")
custom_range.click()


# Initialzing the dates
while True: 
    check_month = driver.find_element(By.XPATH, "/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[2]")
    month = check_month.get_attribute("innerHTML")
    if month == "Aug 2022":
        break
        
    else:
        back_button = driver.find_element(By.XPATH,'/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[1]')

        back_button.click()
        
first_date = driver.find_element(By.XPATH, '/html/body/div[4]/div[3]/div[1]/table/tbody/tr[1]/td[2]')
first_date.click()

#selecting date from Aug to Dec
i=0
while i <= 2:
    front_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/thead/tr[1]/th[3]")
    front_button.click()
    i = i+1

last_date = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/tbody/tr[5]/td[7]")
last_date.click()

apply_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[4]/button[2]")
apply_button.click()
```

```python
#Extracting  the names of vegetables 
for i in range (0,len(final_list)):
    while True:
        try:
            element = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div[1]/div[4]/div/div[1]/div/div[1]/button') # removing deafult vegeatbles presented in search 
            element.click()
        except:
            break
    search_input = driver.find_element(By.NAME, 'search_terms')
    search_input.send_keys(final_list[i]) 
    time.sleep(3)
    search_input.send_keys(Keys.ENTER) #vegetable gets selected 
    
    temp = driver.find_element(By.XPATH, '//*[@id="historyDataTable"]') # extarcting prices 
    time.sleep(6)
    temp_html = temp.get_attribute("innerHTML")
    html_list.append(temp_html)
    time.sleep(6)

#extracting names of vegetables and prices from html code
Name = []
Prices = []

for i in html_list:
    temp_n = re.findall('<tbody><tr><td>(.*?)</td><td>',i)
    Name.extend(temp_n)
    temp_p = re.findall('<td>₹ (.*?)</td>',i)
    Prices.append(temp_p)

#extracting date 
date = re.findall('<th>(.*?)</th>',html_list[0])
dates = []
for i in date:
    temp = re.findall('"white-space: nowrap;">(.*?)</span>',i)
    if len(temp) != 0:
        dates.append(temp)
```

```python
# converting into Dataframe
df = pd.DataFrame(dates,columns = ['Dates'])
df1 = pd.DataFrame(Prices).transpose()
df1.dropna(inplace = True)
df1.columns = [Name]
df = pd.merge(df,df1,left_index=True, right_index=True)
```

```python
#visiting the page
homepage = 'https://vegetablemarketprice.com/'
data = []
html_list = []
driver = webdriver.Firefox()
driver.get(homepage)
search_input = driver.find_element(by=By.XPATH, value='//*[@id="header-search-input"]')

#Searching Noida Price History
header_search_input = 'Noida Vegetable Market Price History'
search_input.send_keys(header_search_input)
search_input.send_keys(Keys.ENTER)
time.sleep(5)
driver.find_element(by=By.XPATH,value='/html/body/ul/li[1]/div').click()

#switching from chart to table in dashboard
switch = driver.find_element(By.XPATH, '//*[@id="navItemTable"]')
switch.click()

#selecting the date 
change_date = driver.find_element(by = By.XPATH,value = '//*[@id="reportrange"]')
change_date.click()
custom_range = driver.find_element(By.XPATH, "//*[contains(text(), 'Custom Range')]")
custom_range.click()

# Initialzing starting date and end date
while True: 
    check_month = driver.find_element(By.XPATH, "/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[2]")
    month = check_month.get_attribute("innerHTML")
    if month == "Jan 2023":
        break
        
    else:
        back_button = driver.find_element(By.XPATH,'/html/body/div[4]/div[2]/div[1]/table/thead/tr[1]/th[1]')

        back_button.click()
        
first_date = driver.find_element(By.XPATH, '/html/body/div[4]/div[2]/div[1]/table/tbody/tr[2]/td[1]')
first_date.click()

i=0                  #switching page of calendar 
while i < 1:
    front_button = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/thead/tr[1]/th[3]")
    front_button.click()
    i = i+1

last_date = driver.find_element(By.XPATH,"/html/body/div[4]/div[3]/div[1]/table/tbody/tr[5]/td[6]")  # selecting end date 
last_date.click()

#applyig the dates 
wait = WebDriverWait(driver, 60)  # Adjust the timeout as needed
apply_button = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[4]/div[4]/button[2]")))
apply_button.click()
```

```python
# extracing the names of vegetables
for i in range (0,len(final_list)):
    while True:
        try:
            element = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div[1]/div[4]/div/div[1]/div/div[1]/button')
            element.click()
        except:
            break
    search_input = driver.find_element(By.NAME, 'search_terms')
    search_input.send_keys(final_list[i])
    time.sleep(3)
    search_input.send_keys(Keys.ENTER)
    
    temp = driver.find_element(By.XPATH, '//*[@id="historyDataTable"]')
    time.sleep(3)
    temp_html = temp.get_attribute("innerHTML")
    html_list.append(temp_html)
    time.sleep(3)
    print(final_list[i])

# extracting the dates from html code
date = re.findall('<th>(.*?)</th>',html_list[0])
dates = []
for i in date:
    temp = re.findall('"white-space: nowrap;">(.*?)</span>',i)
    if len(temp) != 0:
        dates.append(temp)


# Extracting names and prices of the vegetables 
Name = []
Prices = []

for i in html_list:
    temp_n = re.findall('<tbody><tr><td>(.*?)</td><td>',i)
    Name.extend(temp_n)
    
    temp_p = re.findall('<td>₹ (.*?)</td>',i)
    Prices.append(temp_p)
```

```python
#creating a dataframe
df = pd.DataFrame(dates,columns = ['Dates'])
df1 = pd.DataFrame(Prices).transpose()
df1.dropna(inplace = True)
df1.columns = [Name]
df = pd.merge(df,df1,left_index=True, right_index=True)

# exporting csv file 
path = 'E:\\Masai DA Projects\\AN_MARCH_Prices_Noida.csv'
df.to_csv(path, index=False)
```
